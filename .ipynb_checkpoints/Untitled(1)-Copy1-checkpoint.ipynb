```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/out.csv')
```

```python
len(df)
```

```python
df['Date']
```

```python
df['Month'] = df['Date'].apply(lambda x: ((int(x[5:7])-1) * 2) + 1 if int(x[8:]) > 15 else (int(x[5:7])-1) * 2)
```

```python
s = df.groupby('Date')[df.columns[6]].mean()
```

```python
s
```

```python
from matplotlib import pyplot
from statsmodels.tsa.seasonal import seasonal_decompose
series = np.array(s)
result = seasonal_decompose(series, model='additive', freq=30)
result.plot()
pyplot.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(series)
```

```python
model.summary()
```

```python
my_order = (1, 1, 1)
my_seasonal_order = (1, 1, 1, 30)
# define model
model = SARIMAX(series, order=my_order, seasonal_order=my_seasonal_order,enforce_invertibility=False)
```

```python
model_fit = model.fit()
model_fit.forecast(30)
```
